```python
import pandas as pd
```

```python
vine_df = pd.read_csv('vine.csv')
vine_df
```

```python
# filter with total_votes equal or greater than 20
gt20_df = vine_df.loc[vine_df['total_votes'] >= 20]
gt20_df
```

```python
helpful_df = gt20_df.loc[(vine_df['helpful_votes'] / vine_df['total_votes']) >= 0.5]
helpful_df
```

```python
helpful_y_df = helpful_df.loc[helpful_df['vine'] == 'Y']
helpful_y_df
```

```python
helpful_n_df = helpful_df.loc[helpful_df['vine'] == 'N']
helpful_n_df
```

```python
paid_total_five_star_review = helpful_y_df.loc[helpful_y_df['star_rating'] == 5.0].shape[0]
paid_total_review = helpful_y_df.shape[0]
paid_percent = paid_total_five_star_review / paid_total_review

unpaid_total_five_star_review = helpful_n_df.loc[helpful_n_df['star_rating'] == 5.0].shape[0]
unpaid_total_review = helpful_n_df.shape[0]
unpaid_percent = unpaid_total_five_star_review / unpaid_total_review

print("Paid: Five-star: " + str(paid_total_five_star_review) + "; Total: " + str(paid_total_review) + "; Percentage: " + str(paid_percent) + "%")
print("Unpaid: Five-star: " + str(unpaid_total_five_star_review) + "; Total: " + str(unpaid_total_review) + "; Percentage: " + str(unpaid_percent) + "%")
```
